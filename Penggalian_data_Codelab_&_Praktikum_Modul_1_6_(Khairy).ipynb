1. Hapus entri yang tidak lengkap atau memiliki nilai kosong.
2. Hitung jumlah entri yang tersisa setelah penghapusan.

```python
import statistics

matkul_nilai = (
    {"Matematika": 80, "Fisika": 75, "Kimia": "", "Biologi": 85, "Informatika": 90,
     "Sejarah": 70, "Ekonomi": "", "Geografi": 88, "Sosiologi": 92, "Statistika": ""}
)

matkul_nilai_bersih = {k: v for k, v in matkul_nilai.items() if isinstance(v, (int, float))}
print("Data setelah menghapus entri kosong:", matkul_nilai_bersih)

jumlah_tersisa = len(matkul_nilai_bersih)
print("Jumlah entri setelah penghapusan:", jumlah_tersisa)

matkul_sorted = dict(sorted(matkul_nilai_bersih.items(), key=lambda x: x[1]))
print("Data setelah sorting ascending:", matkul_sorted)

matkul_tertinggi = max(matkul_nilai_bersih, key=matkul_nilai_bersih.get)
print("Mata kuliah dengan nilai tertinggi:", matkul_tertinggi)

rata_rata = statistics.mean(matkul_nilai_bersih.values())
print("Rata-rata nilai mata kuliah:", rata_rata)
```

Lakukan analisis berikut pada DataFrame tersebut:
1. Tampilkan ringkasan statistik deskriptif yang mencakup nilai minimum, maksimum, jumlah, rata-rata, standar deviasi, serta nilai kuartil 1, 2, dan 3.
2. Tentukan 3 brand dengan penghasilan terbesar (harga x terjual).
3. Kelompokkan brand dengan jenis penjualan yang sama.
4. Buatlah kolom baru yang berisi kategori ‘Murah’, ‘Sedang’, dan ‘Mahal’. Kategori ini didasarkan pada aturan:
- Harga di bawah kuartil 1 dikategorikan sebagai ‘Murah’
- Harga di atas kuartil 3 dikategorikan sebagai ‘Mahal’
- Sisanya dikategorikan sebagai ‘Sedang’.


```python
import pandas as pd
import statistics

data = {
    "Brand": ["Apple", "Xiaomi", "Samsung", "Infinix", "Oppo", "Asus", "Lenovo", "Huawei", "Google Pixel", "Sony"],
    "Tahun Rilis": [2015, 2017, 2018, 2015, 2017, 2020, 2019, 2019, 2020, 2017],
    "Terjual": [150, 75, 120, 85, 75, 120, 50, 90, 55, 50],
    "Harga (juta)": [19, 9, 10, 9, 4, 9, 15, 15, 15, 11],
    "Jenis Penjualan": ["Online", "Offline", "Online", "Online", "Offline", "Online", "Offline", "Online", "Offline", "Offline"]
}
df = pd.DataFrame(data)

print("Ringkasan Statistik:")
print(df.describe())

df["Penghasilan"] = df["Harga (juta)"] * df["Terjual"]
top_brands = df.nlargest(3, "Penghasilan")
print("\n3 Brand dengan Penghasilan Terbesar:")
print(top_brands[["Brand", "Penghasilan"]])

jenis_penjualan_grouped = df.groupby("Jenis Penjualan")["Brand"].apply(list)
print("\nBrand dengan Jenis Penjualan yang Sama:")
print(jenis_penjualan_grouped)

q1 = df["Harga (juta)"].quantile(0.25)
q3 = df["Harga (juta)"].quantile(0.75)

def kategori_harga(harga):
    if harga < q1:
        return "Murah"
    elif harga > q3:
        return "Mahal"
    else:
        return "Sedang"

df["Kategori"] = df["Harga (juta)"].apply(kategori_harga)
print("\nDataFrame dengan Kategori Harga:")
print(df)
```

- Gap Statistics

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import MinMaxScaler

def compute_gap_statistic(X, k_max=10, n_refs=10):
    gaps = []
    s_k = []
    ref_disps = []

    shape = X.shape
    tops = X.max(axis=0)
    bots = X.min(axis=0)
    dists = tops - bots
    rands = np.random.random_sample(size=(n_refs, shape[0], shape[1]))

    for i in range(n_refs):
        rands[i] = rands[i] * dists + bots

    for k in range(1, k_max+1):
        kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
        disp = np.mean(np.min(pairwise_distances(X, kmeans.cluster_centers_), axis=1))

        ref_disp = np.zeros(n_refs)
        for i in range(n_refs):
            kmeans_ref = KMeans(n_clusters=k, random_state=42).fit(rands[i])
            ref_disp[i] = np.mean(np.min(pairwise_distances(rands[i], kmeans_ref.cluster_centers_), axis=1))

        gap = np.log(np.mean(ref_disp)) - np.log(disp)
        sdk = np.std(np.log(ref_disp)) * np.sqrt(1 + 1 / n_refs)

        gaps.append(gap)
        s_k.append(sdk)
        ref_disps.append(np.mean(ref_disp))

    return gaps, s_k

gaps, s_k = compute_gap_statistic(df_scaled, k_max=10)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), gaps, marker='o', linestyle='--', color='b', label='Gap Statistic')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('Gap Value')
plt.title('Gap Statistic untuk Menentukan K Optimal')
plt.grid(True)
plt.legend()
plt.show()

for k, gap in enumerate(gaps, 1):
    print(f'K={k}, Gap={gap:.4f}')
```

| Aspek                  | **K-Means**                                              | **DBSCAN**                                              |
| ---------------------- | -------------------------------------------------------- | ------------------------------------------------------- |
| **Bentuk Cluster**     | Cenderung bulat/sferis (simetris)                        | Bisa bentuk bebas/irreguler                             |
| **Jumlah Cluster**     | Harus ditentukan di awal (misalnya K = 3)                | Ditentukan otomatis berdasarkan kepadatan data          |

| **Sensitivitas**       | Sangat sensitif terhadap **skala fitur** dan **outlier** | Sensitif terhadap **parameter `eps` dan `min_samples`** |
| **Hasil pada dataset** | Mengelompokkan semua pasien ke dalam cluster             | Mengelompokkan pasien padat, dan mengisolasi yang noise |
| **Kecepatan**          | Cepat untuk dataset besar                                | Bisa lebih lambat, terutama jika dimensi data tinggi    |
| **Cocok untuk**        | Data dengan cluster bulat dan distribusi seragam         | Data dengan cluster tidak beraturan atau banyak noise   |

**5. Content-Based Filtering dengan Cosine Similarity**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Name'])

cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend(title, df, cosine_sim, top_n=5):
    try:
        idx = df[df['Name'] == title].index[0]
    except IndexError:
        return f"Judul '{title}' tidak ditemukan."

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    anime_indices = [i[0] for i in sim_scores]

    return df.iloc[anime_indices][['MAL_ID', 'Name', 'Score', 'Genres', 'Episodes', 'Rating']]

recommend("Naruto", df, cos_sim)
```

```python
from IPython.display import display

for judul in ["Naruto", "One Piece", "Trigun", "Final Approach", "School Rumble"]:
    print(f"\nRekomendasi untuk '{judul}':")
    display(recommend(judul, df, cos_sim))
```
